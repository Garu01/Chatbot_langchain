# Using PyMuPDF

```python
import os
from langchain.document_loaders import PyPDFLoader
from langchain.document_loaders import PyMuPDFLoader
import pytesseract
from pdf2image import convert_from_path

# Base path for books
book_paths = "/content/drive/MyDrive/Chat_bot/Data/Book"

# Your categorized PDF dictionary
dict_pdf = {
    "Math": ["Calculus.pdf", "Algebra.pdf", "Linear Algebra and Its Applications (4th Ed).pdf"],
    "Programming language": ["C.pdf", "Python with Example.pdf"],
    "Data": ["Analytic, Data science and AI.pdf", "Book-5Kimball_The-Data-Warehouse-Toolkit-3rd-Edition-5.pdf"],
    "Computer Science": ["Operating System Concepts 10th 2018.pdf", "Introduction.to.Algorithms.4th.Leiserson.Stein.Rivest.Cormen.MIT.Press.9780262046305.EBooksWorld.ir.pdf",
                         "computer-networking-top-down-approach-8th.pdf", "Data Structure and Algorithm.pdf"],
    "Machine Learning": ["Machine Learning.pdf", "Machine learning co ban.pdf"],
    "Robotics": ["robotic.pdf"]
}

# Initialize a list to store all loaded documents with metadata
loaded_documents_pymu = []

# Helper function to perform OCR on a PDF file
def ocr_pdf(pdf_path, category, pdf_file):
    pages = convert_from_path(pdf_path, dpi=300)
    page_texts = []

    for i, page in enumerate(pages):
        text = pytesseract.image_to_string(page)
        loaded_documents.append({
            "category": category,
            "file_name": pdf_file,
            "page_number": i + 1,
            "content": text
        })
    print(f"Processed OCR for '{pdf_file}' under category '{category}'")

# Main function to load PDFs with fallback options
for category, pdf_files in dict_pdf.items():
    for pdf_file in pdf_files:
        pdf_path = os.path.join(book_paths, pdf_file)

        try:
            # First try loading with PyPDFLoader
            loader = PyPDFLoader(pdf_path)
            docs = loader.load()

            # Add metadata to each document
            for doc in docs:
                doc.metadata["category"] = category
                doc.metadata["file_name"] = pdf_file
                loaded_documents_pymu.append(doc)
            print(f"Loaded '{pdf_file}' under category '{category}' using PyPDFLoader")

        except Exception as e:
            print(f"PyPDFLoader failed for '{pdf_file}', trying PyMuPDFLoader: {e}")

            # Fallback to PyMuPDFLoader
            try:
                loader = PyMuPDFLoader(pdf_path)
                docs = loader.load()

                # Add metadata to each document
                for doc in docs:
                    doc.metadata["category"] = category
                    doc.metadata["file_name"] = pdf_file
                    loaded_documents_pymu.append(doc)
                print(f"Loaded '{pdf_file}' under category '{category}' using PyMuPDFLoader")

            except Exception as e:
                print(f"PyMuPDFLoader failed for '{pdf_file}', using OCR as last fallback: {e}")

                # Fallback to OCR
                ocr_pdf(pdf_path, category, pdf_file)

# Display metadata and a content preview for the first few documents
for doc in loaded_documents_pymu[:3]:  # Adjust the range for more or fewer previews
    if isinstance(doc, dict):  # OCR pages are stored as dictionaries
        print(f"Category: {doc['category']}, File: {doc['file_name']}, Page: {doc['page_number']}, Content Preview: {doc['content'][:200]}...\n")
    else:  # PyPDFLoader and PyMuPDFLoader pages are LangChain documents
        print(f"Metadata: {doc.metadata}, Content Preview: {doc.page_content[:200]}...\n")
```

# using PyPDF

```python
from langchain.document_loaders import PyPDFLoader, PyMuPDFLoader
import pytesseract
from pdf2image import convert_from_path
import os
from langchain.document_loaders import PyPDFLoader

# Base path for books
book_paths = "/content/drive/MyDrive/Chat_bot/Data/Book"

# Your categorized PDF dictionary
dict_pdf = {
    "Math": ["Calculus.pdf", "Algebra.pdf", "Linear Algebra and Its Applications (4th Ed).pdf"],
    "Programming language": ["C.pdf", "Python with Example.pdf"],
    "Data": ["Analytic, Data science and AI.pdf", "Book-5Kimball_The-Data-Warehouse-Toolkit-3rd-Edition-5.pdf"],
    "Computer Science": ["Operating System Concepts 10th 2018.pdf", "Introduction.to.Algorithms.4th.Leiserson.Stein.Rivest.Cormen.MIT.Press.9780262046305.EBooksWorld.ir.pdf",
                         "computer-networking-top-down-approach-8th.pdf", "Data Structure and Algorithm.pdf"],
    "Machine Learning": ["Machine Learning.pdf", "Machine learning co ban.pdf"],
    "Robotics": ["robotic.pdf"]
}
# Initialize loaded_documents
loaded_documents = []

for category, pdf_files in dict_pdf.items():
    for pdf_file in pdf_files:
        pdf_path = os.path.join(book_paths, pdf_file)

        try:
            # Try loading with PyPDFLoader
            loader = PyPDFLoader(pdf_path)
            docs = loader.load()

            # Add metadata
            for doc in docs:
                doc.metadata["category"] = category
                doc.metadata["file_name"] = pdf_file
                loaded_documents.append(doc)
            print(f"Loaded '{pdf_file}' under category '{category}' using PyPDFLoader")

        except Exception as e:
            print(f"PyPDFLoader failed for '{pdf_file}', using OCR as fallback: {e}")

            # Convert PDF to images and use OCR
            pages = convert_from_path(pdf_path, dpi=300)
            for i, page in enumerate(pages):
                text = pytesseract.image_to_string(page)
                loaded_documents.append({
                    "category": category,
                    "file_name": pdf_file,
                    "page_number": i + 1,
                    "content": text
                })
            print(f"Processed OCR for '{pdf_file}'")
```

# Vectorize

```python
from langchain_community.embeddings import HuggingFaceEmbeddings

model_name = "sentence-transformers/all-mpnet-base-v2"
model_kwargs = {'device': 'cuda'}
encode_kwargs = {'normalize_embeddings': False}
embeddings = HuggingFaceEmbeddings(
    model_name=model_name,
    model_kwargs=model_kwargs,
    encode_kwargs=encode_kwargs
)
```

## Store pypdf

```python
# chia batchsize
from tqdm import tqdm
from langchain.vectorstores import Chroma
chunk_size = 1024
for i in tqdm(range(0, len(loaded_documents), chunk_size), desc="Processing Batches"):
    documents = loaded_documents[i:i+chunk_size]
    Chroma.from_documents(
        documents=documents,
        embedding=embeddings,
        persist_directory="/content/chroma_db",
    )
```
